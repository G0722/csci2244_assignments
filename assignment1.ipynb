```
import numpy as np
import matplotlib.pyplot as plt
```

```
def roll(rolls):
  s = 0
  for i in range(rolls):
      s += np.random.randint(1,7)
  return s
```

```
def simulate(i,dice):
  total = 0
  for x in range(i):
    total += roll(dice)
  avg = round(total/i, 3)
  return i, avg
```

```
def main(upperbound,dice):
  points = dict()
  for i in range(upperbound):
    point = simulate(i+1,dice)
    points[point[0]] = point[1]
  return points
```

```
def show_plot(upperbound,dice):
  points = main(upperbound,dice)
  fig = plt.figure()
  ax = fig.add_axes([0,0,1,1])
  ax.scatter(points.keys(), points.values(),color = 'b')
  ax.set_xlabel("Trials")
  ax.set_ylabel("Average Roll Sum")
  ax.set_title("Dice Roll Trials")
  plt.show()
```

Testing the Galileo problem. Is 35 or 34 more likely with many rolls of 10 fair 6-sided dice? Well it seems that with more trials, the points converge on the space between 34 and 36, which is roughly 35. This is why 35 is more likely than 36 with many rolls of 10 dice.

```
show_plot(1000,10)
```

Now looking at Problem 1 on the Quiz. Is a sum of 12 or 10 more likely with many rolls of 3 fair 6-sided dice? 

```
show_plot(1000,3)
```
