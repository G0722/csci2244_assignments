NumPy used as a random number generator. Matplotlib's pyplot used to visualize our simiulations

```
import numpy as np
import matplotlib.pyplot as plt
```

Rolls "d" dice and returns the sum of the faces

```
def roll(d):
  s = 0
  for i in range(d):
      s += np.random.randint(1,7)
  return s
```

Performs "i" roll(s) and returns a tuple: (trial, average)

```
def simulate(i,d):
  total = 0
  for x in range(i):
    total += roll(d)
  avg = round(total/i, 3)
  return i, avg
```

Performs 1 to "sims" simulation(s) and stores all the results in a dictionary

```
def main(sims,d):
  points = dict()
  for i in range(sims):
    point = simulate(i+1,d)
    points[point[0]] = point[1]
  return points
```

Gathers the corresponding number of trials and their averages, and scatter plots in pyplot

```
def show_plot(sims,d):
  points = main(sims,d)
  fig = plt.figure()
  ax = fig.add_axes([0,0,1,1])
  ax.scatter(points.keys(), points.values(),color = 'b')
  ax.set_xlabel("Trials")
  ax.set_ylabel("Average Roll Sum")
  ax.set_title(str(d)+"-dice Roll Trials")
  plt.show()
```

Testing the Galileo problem. Is 35 or 34 more likely with many rolls of 10 fair 6-sided dice? Well it seems that with more trials, the points converge on the space between 34 and 36, which is roughly 35. This is why 35 is more likely than 36 with many rolls of 10 dice.

```
show_plot(1000,10)
```

Now looking at Problem 1 on the Quiz. Is a sum of 12 or 10 more likely with many rolls of 3 fair 6-sided dice? The points seem to be converging on 10.5, and 10 is closer than 12 to that number, thus 10 is more likely as a sum than 12 is for 3-dice rolls.

```
show_plot(1000,3)
```
